```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests 



page = requests.get("https://weather.com/weather/tenday/l/San+Francisco+CA?canonicalCityId=dfdaba8cbe3a4d12a8796e1f7b1ccc7174b4b0a2d5ddb1c8566ae9f154fa638c")
soup = BeautifulSoup(page.content,'html.parser')
#print(soup)
print(page)



days_list = soup.find_all('div', class_ = 'DetailsSummary--DetailsSummary--2HluQ DetailsSummary--fadeOnOpen--vFCc_')

list_day = []

for days in days_list:
    day_time = days.h2.text.split()[0]
    list_day.append(day_time)

ld1 = list_day[1:11]

print(len(ld1))



short_des = soup.find_all('div', class_ ='DaypartDetails--Content--hJ52O DaypartDetails--contentGrid--1SWty')

sd = []

for i in short_des:
    s_d = i.p.text
    sd.append(s_d)

sd1 = sd[1:11]

print(len(sd1))

def ftc():

    temp = soup.find_all('div', class_ = 'DailyContent--ConditionSummary--1X5kT')

    t = []

    for i in temp:
        t_ = i.span.text
        t.append(t_)

    #print((t))

    l = t[0::2]
    h = t[1::2]


    lc = []
    hc = []

    for i in range(0,len(l)):
        l[i]= int(l[i][0:2])

    for i in range(0,len(h)):
        h[i]= int(h[i][0:2])

    for i in l:
        c = (round((i - 32)*(5/9),2))
        lc.append(c)

    lc1 = lc[1:11]
    #print(len(lc1))

    for i in h:
        c = (round((i - 32)*(5/9),2))
        hc.append(c)

    hc1 = hc[0:10]
    #print((hc1))



    df = pd.DataFrame({ 'Day': ld1,
                            'Description': sd1,
                            'Day-Temp': hc1,
                            'Night-Temp': lc1})

    df['Date'] = pd.date_range(start='02/08/2022', periods=len(df), freq='D')

    return df[['Day','Date', 'Day-Temp', 'Night-Temp', 'Description']]

ftc()

```
